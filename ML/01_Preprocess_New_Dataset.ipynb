# Data Preprocessing for Disease-Symptom Dataset
Prepares the dataset for machine learning model training

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
import joblib
import os
```

```python
# Load dataset
df = pd.read_csv('../ML/Diseases_and_Symptoms_dataset.csv')

print(f'Dataset shape: {df.shape}')
print(f'\nFirst few rows:')
print(df.head())
print(f'\nColumn names (first 10):')
print(df.columns[:10].tolist())
print(f'\nDisease value counts:')
print(df['diseases'].value_counts())
```

```python
# Clean data
# Remove duplicates
df = df.drop_duplicates()
print(f'After removing duplicates: {df.shape}')

# Separate features and target
X = df.drop('diseases', axis=1).values
y = df['diseases'].values

print(f'\nFeature matrix shape: {X.shape}')
print(f'Number of diseases: {len(np.unique(y))}')
print(f'Unique diseases: {list(np.unique(y))[:10]}')
```

```python
# Encode disease labels
le = LabelEncoder()
y_encoded = le.fit_transform(y)

print(f'Disease encoding:')
for i, disease in enumerate(le.classes_):
    print(f'  {i}: {disease}')
```

```python
# Normalize/Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print(f'Scaled features shape: {X_scaled.shape}')
print(f'Mean of scaled features: {X_scaled.mean():.6f}')
print(f'Std of scaled features: {X_scaled.std():.6f}')
```

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)

print(f'Training set: {X_train.shape[0]} samples')
print(f'Test set: {X_test.shape[0]} samples')
print(f'Feature count: {X_train.shape[1]}')
```

```python
# Save preprocessed data and encoders
os.makedirs('../Data/processed', exist_ok=True)
os.makedirs('../Models', exist_ok=True)

np.save('../Data/processed/X_train.npy', X_train)
np.save('../Data/processed/X_test.npy', X_test)
np.save('../Data/processed/y_train.npy', y_train)
np.save('../Data/processed/y_test.npy', y_test)

joblib.dump(le, '../Models/disease_encoder.pkl')
joblib.dump(scaler, '../Models/scaler.pkl')

# Also save symptom names for API
symptom_names = [col for col in df.columns if col != 'diseases']
joblib.dump(symptom_names, '../Models/symptom_names.pkl')

print('✓ All files saved successfully!')
print(f'  - X_train.npy, X_test.npy, y_train.npy, y_test.npy')
print(f'  - disease_encoder.pkl')
print(f'  - scaler.pkl')
print(f'  - symptom_names.pkl ({len(symptom_names)} symptoms)')
```
